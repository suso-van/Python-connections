```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the CLEAN data (Created by Person A)
try:
    df = pd.read_csv('../data/cleaned_crime_data.csv')
    print(f"âœ… Data Loaded! Analyzing {len(df)} rows.")
except FileNotFoundError:
    print("âŒ Error: cleaned_crime_data.csv not found. Did Person A run their code?")

# 2. Setup Visual Style
sns.set_theme(style="whitegrid")

# 3. Visualization 1: Crimes by Hour of Day
plt.figure(figsize=(10, 6))
# We count the rows for each hour
sns.countplot(data=df, x='Hour', palette='viridis')
plt.title('Crime Frequency by Hour of Day', fontsize=15)
plt.xlabel('Hour (0-23)')
plt.ylabel('Number of Crimes')
plt.show()

# 4. Visualization 2: Top 10 Crimes
plt.figure(figsize=(12, 6))
# Get top 10 most frequent crimes
top_crimes = df['Primary Type'].value_counts().head(10).index
sns.countplot(data=df, y='Primary Type', order=top_crimes, palette='magma')
plt.title('Top 10 Most Common Crimes', fontsize=15)
plt.show()
```
