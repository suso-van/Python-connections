```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/insurance.csv')
df.drop_duplicates(inplace=True)
```

```python
print(f"Original Skewness: {df['charges'].skew():.2f}")

plt.figure(figsize=(10, 5))
sns.histplot(df['charges'], kde=True)
plt.title("Original Charges Distribution (Right Skewed)")
plt.show()
```

```python
# Create a new feature: 'Obese' (BMI >= 30)
df['obese'] = (df['bmi'] >= 30).astype(int)

# Check interaction: Obese AND Smoker
df['obese_smoker'] = df['obese'] * (df['smoker'] == 'yes').astype(int)

plt.figure(figsize=(10, 6))
sns.boxplot(x='obese_smoker', y='charges', data=df)
plt.title('Impact of Being Obese AND a Smoker (1 = Yes, 0 = No)')
plt.show()
```
